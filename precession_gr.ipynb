## Precession of the perihelion of Mercury's orbit -- General Relativistic correction

We are going to derive the effect of GR on Mercury's orbit by modelling the Sun-Mercury system in Schwarzschild geometry.  The Sun's mass will define the geometry of spacetime outside it, and Mercury will be the test particle simply following force-free geodesic motion in that geometry.
This treatment is reasonable because:
<ol>
<li>Mass of Sun is about $10^7$ times mass of Mercury, so Mercury can effectively be treated like a test particle w.r.t. a stationary Sun</li>
<li>Schwarzschild Radius ($R_s = 2GM / c^2$) of the Sun is about 3000 mt while its radius is $7 * 10^8$ mt, so Schwarzschild radius lies well inside the physical boundary of the Sun, and we assume Schwarzschild geometry outside the Sun</li>
<li>Average orbital radius of Mercury $<R>$ is $57 * 10^9$ mt, so $R_s / <R>$ is approximately $5 * 10^{-7}$, and so, while the GR effect will be small, for Mercury it can still be measured over a few thousand revolutions due to its cumulative effect</li>
</ol>

\end{array}
\right)\\

Note:
* We are using the $(-1, 1, 1, 1)$ signature on the metric
* We are using coordinates $(t, r, \theta, \phi)$ where the 3 spatial coordinates work like the usual spherical-polar coordinates with the extra assumption that the coordinate system is invalid as $r \rightarrow R_s$ and the geometry tends to flat 3-space Euclidean geometry as $r \rightarrow \infty$
* The metric is spherically symmetric, depends only on $r$ and $\theta$, and is independent of $t$ (i.e. metric is stationary) and $\phi$ (i.e. metric has rotational symmetry about an axis perpendicular to the equatorial plane)

The coordinates $(t, r, \theta, \phi)$ of any particle moving in this geometry are scalar functions of proper-time $\tau$ of the particle as:
$$x^{i }(\tau ) = (t(\tau), r(\tau), \theta(\tau), \phi(\tau))$$

Then, from basic GR theory, the 4 geodesic equations, or EOMs of Mercury, are given by:
$$\frac{d^2x^i}{d\tau ^2} + \Gamma ^i{}_{j k} \frac{dx^j}{d\tau } \frac{dx^k}{d\tau } = 0$$

Writing 4-velocity $U^i = \frac{dx^i}{d\tau }$, these EOMs can be written as:
$$\frac{dU^i}{d\tau } = - \Gamma ^i{}_{j k} U^j U^k$$

Then, substituting the values of $\Gamma$ shown above the 4 EOMs above can be expanded as:

$$\frac{dU^0}{d\tau} = -\frac{2 G M U^0 U^1}{r \left(-2 G M + c^2 r\right)}$$

$$\frac{dU^1}{d\tau} = {U^0}^2 \frac{G M \left(2 G M-c^2 r\right)}{c^4 r^3} - {U^1}^2 \frac{G M}{2 G M r-c^2 r^2} - {U^2}^2 \left(\frac{2 G M}{c^2}-r\right) - {U^3}^2 \frac{\left(2 G M-c^2 r\right) {sin^2}\theta}{c^2}$$

$$\frac{dU^2}{d\tau} = -\frac{2 {U^1}{U^2}}{r} + {U^3}^2 cos\theta sin\theta$$

$$\frac{dU^3}{d\tau} = -2 {U^2} {U^3} cot\theta - 2 \frac{{U^1} {U^3}}{r}$$

Next, if we reorient our coordinate system such $U^2 = 0$ and $\theta = \pi/2$ (i.e. initial motion is on the equatorial plane), then the expanded equations above become:

$$\text{Eqn #1 :} \frac{dU^0}{d\tau} = -\frac{2 G M {U^0} {U^1}}{r \left(-2 G M + c^2 r\right)}$$

$$\text{Eqn #2 :} \frac{dU^1}{d\tau} = {U^0}^2 \frac{G M \left(2 G M - c^2 r\right)}{c^4 r^3} - {U^1}^2 \frac{G M}{2G M r - c^2 r^2} - {U^3}^2 \frac{\left(2 G M - c^2 r\right)}{c^2}$$

$$\text{Eqn #3 :} \frac{dU^2}{d\tau} = 0$$

$$\text{Eqn #4 :} \frac{dU^3}{d\tau} = -\frac{2 {U^1} {U^3}}{r}$$

Note that Eqn #3 implies that if $U^2$ is 0 to start with, it remains 0 and the particle remains on the equatorial
plane forever.  The remaining 3 equations are the EOMs, but these equations are in terms of the 4-velocity $U^i$ and its derivatives $\frac{dU^i}{d\tau}$.

To get equations of motion that we can program for evolution into a computer we will need to convert these to equations in terms of the 3-velocity $v^i = (\frac{dr}{dt}, \frac{d\theta}{dt}, \frac{d\phi}{dt})$ and its derivatives $(\frac{dv^r}{dt}, \frac{dv^{\theta}}{dt}, \frac{dv^{\phi}}{dt})$.<br>

To manipulate the equations we will also need the following auxiliary equations:

$$\text{Eqn #5a :} \frac{dr}{dt} = \frac{dr}{d\tau} / \frac{dt}{d\tau} = U^1 / U^0$$
$$\text{Eqn #5b :} \frac{d\theta}{dt} = \frac{d\theta}{d\tau} / \frac{dt}{d\tau} = U^2 / U^0 = 0 \text{ (since U2 is always zero)}$$
$$\text{Eqn #5c :} \frac{d\phi}{dt} = \frac{d\phi}{d\tau} / \frac{dt}{d\tau} = U^3 / U^0$$

$$\text{Eqn #6 :} \frac{d^2r}{d\tau ^2} = \frac{d}{d\tau}(\frac{dr}{d\tau}) = \frac{d}{d\tau }(\frac{dr}{dt}\frac{dt}{d\tau}) = \frac{d^2r}{dt^2}{(\frac{dt}{d\tau})}^{2} + \frac{dr}{dt} \frac{d^2t}{d\tau ^2} \implies \frac{dU^1}{d\tau} = \frac{d^2r}{dt^2} {U^0}^2 + \frac{dr}{dt} \frac{dU^0}{d\tau}$$

$$\text{Eqn #7 :} \frac{d^2\phi}{d\tau ^2} = \frac{d}{d\tau}(\frac{d\phi}{d\tau }) = \frac{d}{d\tau }(\frac{d\phi }{dt}\frac{dt}{d\tau}) = \frac{d^2\phi }{dt^2}{(\frac{dt}{d\tau })}{2} + \frac{d\phi }{dt} \frac{d^2t}{d\tau ^2} \implies \frac{dU^3}{d\tau} = \frac{d^2\phi }{dt^2} {U^0}^2 + \frac{d\phi }{dt} \frac{dU^0}{d\tau}$$

Then, finally, we can derive the EOMs we are looking for as follows:
* Solving Eqn #1, Eqn #2, and Eqn #6, using Eqns #5 to eliminate $U^i/U^0$ terms, we get:

$$\text{Eqn #8a :} v^r = \frac{dr}{dt}$$

$$\text{Eqn #8b :} \frac{dv^r}{dt} = \frac{-GM}{r^2} + r (\frac{d\phi}{dt})^2 - \left(2\frac{GM}{c^2}\right) (\frac{d\phi}{dt})^2 + (GM) \left(2\frac{GM}{c^2}\right) \frac{1}{r^3} - \left(3\frac{GM}{c^2}\right) (\frac{dr}{dt})^2 \frac{1}{r \left(2\frac{GM}{c^2} - r\right)}$$

* Solving Eqn #1, Eqn #4, and Eqn #7, using Eqns #5 to eliminate $U^i/U^0$ terms, we get:

$$\text{Eqn #9a :} v^{\phi} = \omega = \frac{d\phi}{dt}$$

$$\text{Eqn #9b :} \frac{dv^{\phi}}{dt} = \frac{d\omega}{dt} = -2 (\frac{dr}{dt}) (\frac{d\phi}{dt}) \left(3\frac{GM}{c^2} - r\right) \frac{1}{r \left(2\frac{GM}{c^2} - r\right)}$$

<b>Eqns #8a, #8b, #9a, and #9b are the 4 first order ODEs we can evolve on a computer.</b>

Finally, a few notes about these EOMs:
<ol>
<li>If c $\rightarrow $ $\infty $, Eqns #8b and #9b give back the Newtonian equations of motion of Mercury going round the Sun, which should be reassuring! </li>
